```python
import pandas as pd
import numpy as np
# Column names according to the UCI dataset

columns = [
    "ID", "Diagnosis",
    "Radius_Mean", "Texture_Mean", "Perimeter_Mean", "Area_Mean", "Smoothness_Mean",
    "Compactness_Mean", "Concavity_Mean", "Concave_Points_Mean", "Symmetry_Mean", "Fractal_Dimension_Mean",

    "Radius_SE", "Texture_SE", "Perimeter_SE", "Area_SE", "Smoothness_SE",
    "Compactness_SE", "Concavity_SE", "Concave_Points_SE", "Symmetry_SE", "Fractal_Dimension_SE",

    "Radius_Worst", "Texture_Worst", "Perimeter_Worst", "Area_Worst", "Smoothness_Worst",
    "Compactness_Worst", "Concavity_Worst", "Concave_Points_Worst", "Symmetry_Worst", "Fractal_Dimension_Worst"
]
# Load Dataset

df = pd.read_csv("wdbc.data", header=None, names=columns)

# Display first five rows
df.head()
```

```python
# Shape of dataset
print("Shape of Dataset:")
print(df.shape)
# Information about dataset
print("\nDataset Information:")
print(df.info())
# Statistical Summary
print("\nStatistical Summary:")
print(df.describe())
# Check Data Types
print("\nData Types:")
print(df.dtypes)
# Number of Benign and Malignant Tumors
print("\nDiagnosis Counts:")
print(df["Diagnosis"].value_counts())
```

```python
# Missing Values

print("Missing Values:")
print(df.isnull().sum())
df = df.drop(columns=["ID"], errors="ignore")
df["Diagnosis"] = df["Diagnosis"].map({
    "M":1,
    "B":0
})
X = df.drop("Diagnosis", axis=1)

y = df["Diagnosis"]
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
print("Training Samples:", X_train.shape)
print("Testing Samples:", X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(X_train, y_train)
log_predictions = log_model.predict(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)
knn_predictions = knn_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
print("LOGISTIC REGRESSION RESULTS")
print("Accuracy :", accuracy_score(y_test, log_predictions))
print("Precision :", precision_score(y_test, log_predictions))
print("Recall :", recall_score(y_test, log_predictions))
print("F1 Score :", f1_score(y_test, log_predictions))
print("Confusion Matrix")
print(confusion_matrix(y_test, log_predictions))
```

```python
comparison = pd.DataFrame({
    "Metric":[
        "Accuracy",
        "Precision",
        "Recall",
        "F1 Score"
    ],

    "Logistic Regression":[
        accuracy_score(y_test, log_predictions),
        precision_score(y_test, log_predictions),
        recall_score(y_test, log_predictions),
        f1_score(y_test, log_predictions)
    ],

    "KNN":[
        accuracy_score(y_test, knn_predictions),
        precision_score(y_test, knn_predictions),
        recall_score(y_test, knn_predictions),
        f1_score(y_test, knn_predictions)
    ]
})

print(comparison)
```

```python
log_accuracy = accuracy_score(y_test, log_predictions)
knn_accuracy = accuracy_score(y_test, knn_predictions)

print("\nINTERPRETATION")

if log_accuracy > knn_accuracy:
    print("Logistic Regression performed better than KNN based on Accuracy.")
elif knn_accuracy > log_accuracy:
    print("KNN performed better than Logistic Regression based on Accuracy.")
else:
    print("Both models performed equally well based on Accuracy.")
```
